```python
# Eksik veriler
eksik_veriler = pd.read_csv("../Docs/eksikveriler.csv")
eksik_veriler.isnull().sum().sum()
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = "mean")
yas = eksik_veriler.iloc[:,1:4].values

# Fit ile √∂ƒürenme
imputer.fit(yas[:,1:4])

# Transform ile uygulama
yas[:,1:4] = imputer.transform(yas[:,1:4])
print(yas)
```

```python
# Kategorik veriler
veriler = pd.read_csv("../Docs/veriler.csv")
ulke = veriler.iloc[:,0:1].values
print(ulke)
```

```python
le = preprocessing.LabelEncoder()
ulke[:,0] = le.fit_transform(veriler.iloc[:,0])
print(ulke)
```

```python
# One Hot Encoding
ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke).toarray()
print(ulke)
```

```python
# Verilerin birle≈ütirilmesi ve DataFrame Olu≈üturulmasƒ±
sonuc = pd.DataFrame(data = ulke, index = range(22), columns = ["fr","tr","us"])
sonuc
```
